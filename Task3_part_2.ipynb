```python

#step 1  exploration
import pandas as pd

# Load dataset
df = pd.read_csv("/content/sample_data/dataset-2.csv")

# Basic exploration
print(df.head())
print(df.info())
print(df.describe())
```

```python
# Fill missing Age values with the median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill missing Embarked values with the mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop Cabin column only if it exists
if 'Cabin' in df.columns:
    df.drop(columns=['Cabin'], inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Survival rate by gender
gender_survival = df.groupby('Sex')['Survived'].mean()
# Survival rate by class
class_survival = df.groupby('Pclass')['Survived'].mean()
# Average age per class
avg_age_class = df.groupby('Pclass')['Age'].mean()
# Survival rate by age group
df['AgeGroup'] = pd.cut(df['Age'], bins=[0,12,18,40,60,80],
                        labels=['Child','Teen','Adult','Middle‑aged','Senior'])
agegroup_survival = df.groupby('AgeGroup')['Survived'].mean()
```

```python
# Female passengers who survived
female_survivors = df[(df['Sex']=='female') & (df['Survived']==1)]

# Children who survived
child_survivors = df[(df['Age']<=12) & (df['Survived']==1)]

# 1st‑class passengers with high survival probability
first_class_high = df[(df['Pclass']==1) & (df['Survived']==1)]
```

## Step 5: Insights

- **Who was more likely to survive?**  
  Females had a much higher survival rate (74%) compared to males (19%).

- **Did class affect survival?**  
  Yes. First‑class passengers had the highest survival rate (63%), while third‑class passengers had the lowest (24%).

- **Were children prioritized?**  
  Children (≤12 years) had a higher survival rate (50%) than adults, showing they were given priority.

- **What combination had the highest survival rate?**  
  Female passengers in first class had the greatest chance of survival — around **95%**.
